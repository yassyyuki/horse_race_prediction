```python
#https://db.netkeiba.com/jockey/00666/
#騎手情報のスクレイピング
import pandas as pd
import numpy as np
import requests
from lxml import html
from bs4 import BeautifulSoup
import re

URL = 'https://db.netkeiba.com/jockey/00666/'
HTML = requests.get(URL).content
soup = BeautifulSoup(HTML, 'html.parser')
dfs = pd.read_html(URL)

def get_person_name(soup):
    return soup.find('div',attrs={'class':'Name'}).find('h1').text

def extract_kanji_katakana(name:str):
    kanji_katakana_list = re.findall(r'[\u4e00-\u9fff|\u30a0-\u30ff]+', name)
    return ''.join(kanji_katakana_list)

def set_first_row_as_column(df):
    df_ = df.copy()
    cols = list(df_.iloc[0,:])
    df.columns = cols
    return df.iloc[1:,:]

jockey_name = get_person_name(soup)
jockey_name = extract_kanji_katakana(jockey_name)
jockey_profile = set_first_row_as_column(dfs[0].T)
jockey_profile['name'] = jockey_name
```
